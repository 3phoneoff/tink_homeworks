```python
import numpy as np
import statsmodels.stats.power as smp
import matplotlib.pyplot as plt
```

```python
alpha = 0.05
N = 1000
```

## Размер эффекта 0.1

```python
effect_size = 0.1
p1 = []
for i in range(1, N):
    p1.append(smp.zt_ind_solve_power(effect_size=effect_size, nobs1=i, alpha=alpha, alternative='two-sided'))
    
plt.figure(figsize=(12, 8))
plt.scatter(range(1, N), p1)
plt.xlabel('Размер выборки', fontsize=12)
plt.ylabel('Мощность', fontsize=12)
plt.show()
```

Для сравнения покажем, когда мощность достигает единицы:

```python
effect_size = 0.1
p11 = []
for i in range(1, 6000):
    p11.append(smp.zt_ind_solve_power(effect_size=effect_size, nobs1=i, alpha=alpha, alternative='two-sided'))
    
plt.figure(figsize=(12, 8))
plt.scatter(range(1, 6000), p11)
plt.xlabel('Размер выборки', fontsize=12)
plt.ylabel('Мощность', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(range(1, N), p1)
plt.scatter(range(1, N), p2)
plt.scatter(range(1, N), p3)
plt.xlabel('Размер выборки', fontsize=12)
plt.ylabel('Мощность', fontsize=12)
plt.show()
```
